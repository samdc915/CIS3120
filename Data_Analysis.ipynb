```python
from pandas_datareader import data, wb
```

```python
gdp_data = wb.download(indicator='NY.GDP.PCAP.KD', country=['US', 'CA', 'MX', 'JP', 'PH', 'KR'], start=1980, end=2015)
gdp_data.columns=['GDP']
print(gdp_data)
type(gdp_data)
```

```python
new_data=gdp_data['GDP'].groupby(level=0).mean()
new_data
```

```python
gdp_data.describe()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ax = new_data.plot(kind='bar', title ="GDP of Select Countries", figsize=(15, 10), legend=True, fontsize=18, color='green')
ax.set_xlabel("Country", fontsize=18)
ax.set_ylabel("GDP", fontsize=18)
plt.show()
```
